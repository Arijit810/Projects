```python
import numpy as np
import pandas as pd
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
data = pd.read_csv("train_titanic.csv",delimiter=",")
testing = pd.read_csv("test_titanic.csv",delimiter =",")
data
```

```python
data.drop(["Name","Ticket"],axis=1,inplace=True)
```

```python
def male_ohe(g):
    if g == "male":
        return 1
    else:
        return 0

def female_ohe(g):
    if g == "female":
        return 1
    else:
        return 0
```

```python
data["Male"] = data["Sex"].apply(male_ohe)
data["Female"] = data["Sex"].apply(female_ohe)
```

```python
data.drop("Sex",axis=1,inplace=True)
data.head()
```

```python
data.isnull().sum()
```

```python
def cherbourg(s):
    if s=="C":
        return 1
    else:
        return 0
    
def queenstown(s):
    if s=="Q":
        return 1
    else:
        return 0

def southamptom(s):
    if s=="S":
        return 1
    else:
        return 0
```

```python
data["Cherbourg"] = data.Embarked.apply(cherbourg)
data["Queenstown"] = data.Embarked.apply(queenstown)
data["Southamptom"] = data.Embarked.apply(southamptom)
data.drop("Embarked",axis=1, inplace=True)
data.head()
```

```python
def cabin(x):
    if x is np.nan:
        return 0
    else:
        return 1
```

```python
data["Cabin"] = data.Cabin.apply(cabin)
data.head()
```

```python
data.isnull().sum()
```

```python
data["Age"].fillna(data["Age"].mean(),inplace=True)
data.head()
```

```python
data.describe()
```

```python
data = data[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare','Cabin','Male','Female','Cherbourg','Queenstown','Southamptom','Survived']]
data.head()
```

```python
data = data.values
```

```python
data.shape
```

```python
X = data[:,0:11]
Y = data[:,11].reshape(-1,1)
X.shape,Y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y)
```

```python
clf=LogisticRegression(n_jobs=-1, solver="saga", max_iter=20000000, C=100, tol=0.000001, class_weight="balanced" , random_state = 1)
clf.fit  (x_train, y_train)
clf.score(x_test, y_test)
```

```python
y_test_pred = clf.predict(x_test)
y_train_pred = clf.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train,y_train_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report (y_test,y_test_pred))
```
